```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dt = pd.read_csv("guns.csv", index_col = 0)
dt.shape
```

```python
dt.head()
```

```python
#dt.groupby('year').year.count()
dt.year.value_counts()
```

```python
pal = sns.color_palette("Set2")
bs = dt.intent.value_counts()
my_plot = bs.plot(kind='barh',stacked=True,title="Leading cause of gun death",color=pal, alpha=0.6,figsize=(10, 10))
my_plot.set_xlabel("Number of deaths")
my_plot.set_ylabel("Causes of death")
```

```python
pd.crosstab(dt.education, dt.intent).plot(kind='bar', color=pal, alpha=0.6,figsize=(10, 10))
plt.title('Education Distribution by Intent')
plt.xlabel('Education')
plt.ylabel('Frequency')
plt.legend(loc=0)
sns.despine()
```

```python
pd.crosstab(dt.sex, dt.intent).plot(kind='bar', color=pal, alpha=0.6,figsize=(10, 10))
plt.title('Gender Distribution by Intent')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.legend(loc=0)
sns.despine()
```

```python
pal = sns.color_palette("Set2")
sd  = dt.race.value_counts()
sd.plot(kind='pie', subplots=True, figsize=(10, 10),colors = pal,autopct='%1.1f%%')
#plt.legend(loc='best')
```
